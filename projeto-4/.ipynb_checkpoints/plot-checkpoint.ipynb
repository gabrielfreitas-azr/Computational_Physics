```python
import matplotlib.pyplot as plt
import numpy as np
import scienceplots

plt.style.use(['science', 'notebook'])
```

```python
plt.figure(figsize = (12, 8))

data = np.genfromtxt('tarefa-1/tarefa-1a-saida-1.dat', names = True)

plt.plot(data['tau'], data['theta'], label = '$\\theta$', color = 'black')
plt.plot(data['tau'], data['omega'], label = '$\\omega$', color = 'grey')
plt.plot(data['tau'], data['epsilon'], label = '$\\epsilon$', color = 'gold')

plt.plot(data['tau'], data['theta_i'], label = '$\\theta$ - Euler', color = 'black', ls = '--')
plt.plot(data['tau'], data['omega_i'], label = '$\\omega$ - Euler', color = 'grey', ls = '--' )
plt.plot(data['tau'], data['epsilon_i'], label = '$\\epsilon$ - Euler', color = 'gold', ls = '--' )


plt.xlabel('$\\tau$', fontsize = 20)

plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
```

```python
plt.figure(figsize = (12, 8))

data = np.genfromtxt('tarefa-1/tarefa-1b-saida-1.dat', names = True)

plt.plot(data['tau'], data['theta'], label = '$\\theta$', color = 'black')
plt.plot(data['tau'], data['omega'], label = '$\\omega$', color = 'grey')
plt.plot(data['tau'], data['epsilon'], label = '$\\epsilon$', color = 'gold')

plt.plot(data['tau'], data['theta_i'], label = '$\\theta$ - Euler-Cromer', color = 'black', ls = '--')
plt.plot(data['tau'], data['omega_i'], label = '$\\omega$ - Euler-Cromer', color = 'grey', ls = '--' )
plt.plot(data['tau'], data['epsilon_i'], label = '$\\epsilon$ - Euler-Cromer', color = 'gold', ls = '--' )


plt.xlabel('$\\tau$', fontsize = 20)
plt.ylim(-0.75, 0.75)

plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
```
