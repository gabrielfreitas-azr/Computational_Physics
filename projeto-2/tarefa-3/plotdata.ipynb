```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from matplotlib import colors
```

```python
def aux_func(x, y, ax, ax_histx, ax_histy, binwidth, N):
        
    ax_histx.tick_params(axis="x", labelbottom = False, labelleft = False)
    ax_histy.tick_params(axis="y", labelleft=False)

    xymax = max(np.max(np.abs(x)), np.max(np.abs(y)))
    lim = (int(xymax/binwidth) + 1) * binwidth

    bins = np.arange(-lim, lim + binwidth, binwidth)
    
    ax_histx.hist(x, bins=bins, color = "rebeccapurple", histtype='stepfilled', alpha=0.8, edgecolor="black")
    ax_histy.hist(y, bins=bins, orientation='horizontal', color = "rebeccapurple", histtype='stepfilled', alpha=0.8, edgecolor="black" )
    
    ax_histx.set_yticks([])
    ax_histy.set_xticks([])  
        
    ax_histx.grid(linestyle = '--', linewidth = 0.5)
    ax_histy.grid(linestyle = '--', linewidth = 0.5)

    ax_histx.set_title("Histograma para N = " + str(N))
    ax_histx.set_ylabel("n de and.", fontsize = 10)
    ax_histy.set_xlabel("n de and.", fontsize = 10)
    
    # Hist2d
    
    ax.hist2d(x, y, bins = bins, cmap = "inferno")
    ax.set_xlabel("Posição x [u.a.]")
    ax.set_ylabel("Posição y [u.a.]")

    
def hist2d_with_margin(x, y, bins, N):
    
    # Start with a square Figure.

    fig = plt.figure(figsize=(6, 6))


    # Add a gridspec with two rows and two columns and a ratio of 1 to 4 between
    # the size of the marginal axes and the main axes in both directions.
    # Also adjust the subplot parameters for a square plot.
    
    gs = fig.add_gridspec(2, 2,  width_ratios=(4, 1), height_ratios=(1, 4),
                          left=0.1, right=0.9, bottom=0.1, top=0.9,
                          wspace=0.05, hspace=0.05)

    # Create the Axes.
    ax = fig.add_subplot(gs[1, 0])
    ax_histx = fig.add_subplot(gs[0, 0], sharex = ax)
    ax_histy = fig.add_subplot(gs[1, 1], sharey = ax)

    # Draw the scatter plot and marginals.
    aux_func(x, y, ax, ax_histx, ax_histy, bins, N)
```

# Para N = 10

```python
data = np.genfromtxt("tarefa-3-saida-1.dat")

x = []
y = []

for xy in data:
    x.append(xy[0])
    y.append(xy[1])
    
hist2d_with_margin(x, y, 2, 10)

plt.savefig("tarefa-3-grafico-1.png", facecolor='white', transparent=False, bbox_inches = 'tight')
```
