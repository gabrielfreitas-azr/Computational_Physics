```python
fig = plt.figure(figsize = (16, 8))

data = np.genfromtxt('tarefa-1/tarefa-1c/tarefa-1c-saida-1.dat', names = True)
data1 = np.genfromtxt('tarefa-1/tarefa-1c/tarefa-1c-saida-2.dat', names = True)

plt.subplot(1, 2, 1)
plt.plot(data['dtau'], data['delta'], color = 'navy')
plt.xlabel('$\\Delta \\tau$', loc = 'right')
plt.ylabel('$delta $ $(\\delta)$', loc = 'top')
plt.xlim(0.105, -0.005)

plt.subplot(1, 2, 2)
plt.plot(data1['dtau'][1:-1], data1['delta'][1:-1], color = 'navy')
plt.xlabel('$\\Delta \\tau$', loc = 'right')
plt.ylabel('$delta $ $(\\delta)$', loc = 'top')
plt.xlim(0.105, 0)
         
print(data['dtau'][-1])
print(data1['dtau'][-1])
```

```python
fig = plt.figure(figsize = (8, 8))

data = np.genfromtxt('tarefa-1/tarefa-1d/tarefa-1d-saida-1.dat', names = True)
data1 = np.genfromtxt('tarefa-1/tarefa-1d/tarefa-1d-saida-2.dat', names = True)

plt.plot(data['tau'], data['E'], color = 'navy', label = 'Euler-Cromer')
plt.plot(data1['tau'], data1['E'], color = 'blue', label = 'Verlet')

plt.legend()

plt.ylim(3, 4.6)
plt.xlabel('$\\tau$')
plt.ylabel('Energia E $\\left(\\frac{GM_sm}{(UA)}\\right)$', loc = 'top')
```
